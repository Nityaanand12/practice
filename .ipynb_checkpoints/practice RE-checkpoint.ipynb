```python
import re
```

```python
matcher = re.finditer('[abc]+','abab1abab a231 1a')
for match in matcher:
    print(match.start(),'....',match.group())
```

```python
matcher = re.finditer('\t','a\tbab1abab a231 1a')
for match in matcher:
    print(match.start(),'....',match.group())
```

```python
s=input("Enter mobile number")
m=re.fullmatch('\d{10,13}',s)
if m:
    if re.fullmatch('\d{10}',s):
        if re.fullmatch('[6-9]\d{9}',s):
            print(s,'is accepted')
        else:
            print("Enter correct mobile number")
    elif re.fullmatch('\d{11}',s):
        if re.fullmatch('0[6-9]\d{9}',s):
            print(s,'is accepted')
        else:
            print("Enter correct mobile number")
    elif re.fullmatch('\d{12}',s):
        if re.fullmatch('91[6-9]\d{9}',s):
            print(s,'is accepted')
        else:
            print("Enter correct mobile number")
    elif re.fullmatch('\d{13}',s):
        if re.fullmatch('+91[6-9]\d{9}',s):
            print(s,'is accepted')
        else:
            print("Enter mobile number correctly")
else:
    print('Enter the number correctly')
    
```
