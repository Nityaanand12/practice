```python
q = "create table customer(Id int,Name varchar(255),Age int,Salary number,Location varchar(255),Skillset varchar(255),Deg varchar(255))"
cur.execute(q)
```

```python
fp = open('data1.csv')
content = fp.readlines()
fp.close()
content.pop(0)
#print(content)
res_data = [ldata.strip('\n').split(';') for ldata in content]
for res in res_data:
    q="insert into customer(Id,Name,Age,Salary,Location,Skillset,Deg) values(%s,'%s',%s,%s,'%s','%s','%s')" %(res)
    print(tuple(res))
    cur.execute(q)
    con.commit()
```

```python
q = "select Name,age,count(*) as c from customer group by Name,age order by c desc"
select_s = cur.execute(q)
for select in select_s:
    print(select)
```
